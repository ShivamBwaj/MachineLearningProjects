```python
df1=pd.read_excel('Online Retail.xlsx')
```

```python
df1[df1.Description.isnull()]
```

```python
most_freq=df1[['StockCode','Description']].value_counts().reset_index()
most_freq
```

```python
most_freq[most_freq.StockCode=='85123A']
```

```python
most_freq=most_freq.groupby('StockCode').head(1)
```

```python
most_freq
```

```python
most_freq.columns=['StockCode','freq_Description','Count']
df2=df1.merge(most_freq,on='StockCode',how='left')
df2.head()
```

```python
df2['Description']=df2['freq_Description']
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.dropna(subset='Description',inplace=True)
df2.isnull().sum()
```

```python
df2.drop(columns=['freq_Description','Count'],inplace=True)
df2.head()
```

```python
df2.describe()
```

```python
df2[df2.UnitPrice<0]
```

```python
df3=df2[(df2.UnitPrice>0) &(df2.Quantity>0)]
df3.head()
```

```python
df3.describe()
```

```python
df3.Quantity.quantile(0.99)
```

```python
df4=df3.copy()
df4['TotalSales']=df4['Quantity']*df4['UnitPrice']
df4.head()
```

```python
df4['Month']=df4['InvoiceDate'].dt.month
df4.sample(5)
```

```python
monthly_sales=df4.groupby('Month')['TotalSales'].sum()
monthly_sales.plot(kind='bar',title='Sales(Monthly)')
import matplotlib.pyplot as plt
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
monthly_sales.plot(kind='line',title='Sales(Monthly)',marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
hmm=df4.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(5)
hmm.plot(kind='barh')
plt.xlabel("Sales in 10^6")
plt.ylabel("Country")
plt.title("TOP 5 countries")
plt.show()
```

```python
ok=df4.groupby('StockCode')['TotalSales'].sum().sort_values(ascending=False).head(5)
ok.plot(kind='barh',color='skyblue')
plt.show()
ok.plot(kind='line',color='red',marker='o')
plt.show()
```
